# ReSampling, Shifting, Windowing

```python
df = pd.DataFrame({
    #D = day ; W = week
    "date":pd.date_range(start = '2021-03-01',periods = 100, freq = 'D'),
    "temperature":np.random.randint(18,40,size = 100) + np.random.random(100).round(1)
})
```

### Random numbers
no method to create a floating point number in a given range.
 Create a random integer and add it with random number produced by "random.random" function

## Shift
shifts the values or data points to the next (or step value we desire to) row for comparision

```python
df['lag'] = df['temperature'].shift(1)
df
```

###### NaN - Not A Number

```python
df['lag '] = df['temperature'].shift(1,fill_value=df['temperature'].mean())
```

```python
df.temperature.mean()
```

```python
df.head()
```

### Resample

```python
df_weekly = df.resample('W',on = 'date').mean()
df_weekly.head()
```

```python
df_monthly = df.resample('M',on = 'date').mean()
df_monthly
```
